```python
import numpy as np
import math
m=10
n=10
U=100
M=np.zeros((m,n))
for i in range(len(M)):
    for j in range(len(M[i])):
        M[i][j]+=np.random.randint(0, U)
for i in range (len(M)):
        for j in range (len(M[i])):
            if i == 0 :
                M[i][j]=0
            elif i == (len(M)-1) :
                M[i][j]=U
            elif j == 0 or j==(len(M[i])-1) :
                M[i][j]=(U/(len(M)-1))*i
N = np.zeros((m,n))
for x in range(0, len(N)):
    for y in range(0, len(N[x])):
        if x==0:
            N[x][y]=1
        elif y==0:
            N[x][y]=1
        elif x==(len(N)-1):
            N[x][y]=1
        elif y==(len(N[x])-1):
            N[x][y]=1
N1=N
print(M)
print(N)
```

```python
while (np.sum(N)!=n*m):
    for i in range (1,len(M)-1):
        for j in range (1,len(M[i])-1):
            g=M[i][j]
            M[i][j] = (1/4)*(M[i-1][j]+M[i][j-1]+M[i+1][j]+M[i][j+1])
            if (M[i][j]-g<=0.05):
                N[i][j]=1
    print (M)
    print(N)
    N=N1
```
