```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
M0=[3.3,2.2]
pcr=26
N=33
q=1
tau1=25
tau2=14
p1=5.5
p2=11
a1 = pcr/(tau1*tau1*p1*p1*N*q)
a2 = pcr/(tau2*tau2*p2*p2*N*q)
b = pcr/(tau1*tau1*tau2*tau2*p1*p1*p2*p2*N*q)
c1 = (pcr-p1)/(tau1*p1)
c2 = (pcr-p2)/(tau2*p2)
def syst1(M,t):
    dM1=M[0]-(b/c1)*M[0]*M[1]-(a1/c1)*M[0]*M[0]
    dM2=(c2/c1)*M[1]-(b/c1)*M[0]*M[1]-(a2/c1)*M[1]*M[1]
    return([dM1, dM2])
t=arange(0,30,0.01)
y=odeint(syst1,M0,t)
plt.plot(t,y)
plt.legend('12')
plt.show()
```
