```python
from numpy import *
from math import *
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

Первый случай:

```python
w=sqrt(5.2)
g=0
def f(t):
    f=sin(0.00*t)
    return f
def dx(x,t):
    dx1=x[1]
    dx2=-w*w*x[0]-g*x[1]-f(t)
    return [dx1,dx2]
```

```python
t0=0
t=arange(t0,59,0.05)
x0=[0.5,-1.5]
```

```python
x=odeint(dx,x0,t)
y=[[elem[0] for elem in x],[elem[1] for elem in x]]
```

```python
plt.grid()
plt.title('Решение 1')
plt.plot(t,x)
```

```python
def phaze(der, title): #функция получения и вывода фазового портрета
    x = odeint(der,x0,t)
    y1 = x[:,0]
    y2 = x[:,1]
    
    plt.grid()
    plt.title(title)
    plt.yticks(arange(-2,2,0.2))
    plt.xticks(arange(-2,2,0.2))
    plt.ylabel('y')
    plt.xlabel('x')
    plt.plot(y1, y2)
```

```python
phaze(dx, 'Фазовый портрет 1')
```
