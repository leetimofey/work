```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#I(0)<=I*
a=0.01
b=0.02
N=11900
I0=290
R0=52
S0=N-I0-R0
def syst(x,t):
    dS=0
    dI=-b*x[1]
    dR=b*x[1]
    return([dS,dI,dR])
x0=[S0,I0,R0]
t=arange(0,200,0.01)
y=odeint(syst,x0,t)
plt.plot(t,y)
plt.legend('SIR')
plt.show()
```

```python
#I(0)>I*
def syst2(x,t):
    dS=-a*x[0]
    dI=a*x[0]-b*x[1]
    dR=b*x[1]
    return([dS,dI,dR])
x0=[S0,I0,R0]
t=arange(0,400,0.01)
y=odeint(syst2,x0,t)
plt.plot(t,y)
plt.legend('SIR')
plt.show()
```
