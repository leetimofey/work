```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
a=0.25
b=0.45
c=0.025
d=0.045
x0=[8,11]
t=arange(0,400,0.1)
def syst(x,t):
    dx1=-a*x[0]+c*x[0]*x[1]
    dx2=b*x[1]-d*x[0]*x[1]
    return([dx1,dx2])
y=odeint(syst,x0,t)
y1=y[:,0]
y2=y[:,1]
```

```python
plt.title('Популяция хищников')
plt.plot(t,y1)
```

```python
plt.title('Популяция жертв')
plt.plot(t,y2)
```

```python
plt.title('Зависимость популяции хищников от жертв')
plt.plot(y1,y2)
```

```python
xst=[b/d, a/c]
yst=odeint(syst,xst,t)
y1st=yst[:,0]
y2st=yst[:,1]
plt.title('Стационарное состояние системы (хищники):')
plt.plot(t,y1st)
```

```python
plt.title('Стационарное состояние системы (жертвы):')
plt.plot(t,y2st)
```
